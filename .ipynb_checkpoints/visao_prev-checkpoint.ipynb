## Relatório Fictício - Clusterização e Análise de Texto (NPL)

```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import warnings

# Ignorar os warnings
warnings.filterwarnings("ignore")

import os
os.environ["OMP_NUM_THREADS"] = "1"
```

### Definindo dados fictícios

```python
# Definindo os dados fictícios
np.random.seed(42)

clientes = []
for i in range(1, 21):
    for ano in range(2015, 2025):
        idade = np.random.randint(25, 65)
        contribuicao_anual = np.random.randint(1000, 20000)
        saldo_total = contribuicao_anual * np.random.randint(5, 10) + np.random.randint(10000, 100000)
        rentabilidade = np.random.uniform(0.02, 0.1)
        
        clientes.append([f'Cliente_{i}', ano, contribuicao_anual, saldo_total, idade, rentabilidade])

# Criando o DataFrame
df = pd.DataFrame(clientes, columns=['ClienteID', 'Ano', 'Contribuicao_Anual', 'Saldo_Total', 'Idade', 'Rentabilidade'])

# Exibindo as primeiras linhas do DataFrame
df.head()
```

### Clusterziação utilizando K-Means

```python
# Pré-processamento dos dados para clusterização
df_cluster = df.groupby('ClienteID').agg({
    'Contribuicao_Anual': 'mean',
    'Saldo_Total': 'mean',
    'Idade': 'mean',
    'Rentabilidade': 'mean'
}).reset_index()

# Normalizando os dados
scaler = StandardScaler()
df_cluster_scaled = scaler.fit_transform(df_cluster[['Contribuicao_Anual', 'Saldo_Total', 'Idade', 'Rentabilidade']])

# Aplicando o K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
df_cluster['Cluster'] = kmeans.fit_predict(df_cluster_scaled)
```

```python
# Definindo cores para os clusters
colors = {0: 'red', 1: 'green', 2: 'blue'}

# Criando o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(df_cluster['Contribuicao_Anual'], df_cluster['Saldo_Total'], 
            c=df_cluster['Cluster'].map(colors), label=df_cluster['Cluster'], s=100)

plt.title('Clusterização de Clientes de Previdência Privada')
plt.xlabel('Contribuição Anual Média')
plt.ylabel('Saldo Total Médio')
plt.grid(True)

# Adicionando uma legenda
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, [f'Cluster {i}' for i in range(3)], title="Clusters", loc='upper right')

# Exibindo o gráfico
plt.show()
```

Distribuição de Clusters: No gráfico de dispersão, os clientes foram agrupados em três clusters distintos, representados por diferentes cores (azul, verde e vermelho).\
Clusters Identificados:
- O cluster azul parece representar clientes com contribuições anuais mais baixas (em torno de 8.000 a 10.000) e saldos totais médios menores (em torno de 110.000 a 120.000).
- O cluster verde abrange clientes com contribuições anuais médias e saldos totais que variam mais amplamente.
- O cluster vermelho representa clientes com contribuições anuais mais elevadas (próximas a 12.000) e saldos totais mais altos (em torno de 130.000 a 140.000

### Algoritmo Hierárquico (Hierarchical Clustering)

```python
# Calculando o linkage matrix para a clusterização hierárquica
Z = linkage(df_cluster_scaled, method='ward')

# Plotando o dendrograma
plt.figure(figsize=(12, 8))
dendrogram(Z, labels=df_cluster['ClienteID'].values, leaf_rotation=90, leaf_font_size=10)
plt.title('Dendrograma da Clusterização Hierárquica')
plt.xlabel('Clientes')
plt.ylabel('Distância')
plt.show()
```

Se cortarmos o dendrograma em diferentes níveis, podemos identificar o mesmo número de clusters que no K-Means ou até mais, dependendo da granularidade desejada. Por exemplo, podemos observar que há pelo menos três clusters principais (laranja, verde, e vermelho), com o azul sendo o grupo mais distinto.

## Análise de Sentimento

```python
from textblob import TextBlob

# Feedbacks dos clientes
translated_feedbacks = [
    "I am very satisfied with the service and the profitability of my pension plan.",
    "I think the communication could be better. I often get confused with the reports.",
    "The returns on my investment have been consistently good. I recommend the service.",
    "I am not happy with the fees they are charging. They are very high compared to the market.",
    "The service is good, but I feel there could be more investment options.",
    "Excellent! It exceeded my expectations in every aspect.",
    "I am dissatisfied with the lack of transparency in operations. I need more clarity.",
    "Very good! I like the ease of access to information and the return on investment.",
    "The response times for service are too long. I am not satisfied with that."
]

# Função para analisar o sentimento
def analyze_sentiment_adjusted(feedback):
    analysis = TextBlob(feedback)
    if analysis.sentiment.polarity > 0.1:
        return "Positivo"
    elif analysis.sentiment.polarity < -0.1:
        return "Negativo"
    else:
        return "Neutro"
```

```python
# Aplicando a análise de sentimento ajustada em inglês
sentimentos_ajustados_en = [analyze_sentiment_adjusted(feedback) for feedback in translated_feedbacks]

# Contando a frequência de cada sentimento
sentiment_counts = pd.Series(sentimentos_ajustados_en).value_counts()

# Plotando o gráfico de barras
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Distribuição dos Sentimentos dos Feedbacks')
plt.xlabel('Sentimento')
plt.ylabel('Número de Feedbacks')
plt.xticks(rotation=0)
plt.show()
```
